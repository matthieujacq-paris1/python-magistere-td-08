# L'h√©ritage des classes

L'h√©ritage est un concept de programmation orient√©e objet qui permet de d√©finir une classe en utilisant une autre classe comme mod√®le. La classe qui est utilis√©e comme mod√®le est appel√©e la **classe de base** (ou classe parent), et la nouvelle classe est appel√©e la **classe d√©riv√©e** (ou classe enfant). La classe d√©riv√©e h√©rite de toutes les propri√©t√©s et m√©thodes de la classe de base, et peut ajouter de nouvelles propri√©t√©s et m√©thodes.

Pour d√©finir une classe d√©riv√©e √† partir d'une classe de base, on utilise la syntaxe suivante `class NomClasseDerivee(NomClasseDeBase):` :

```python
class BaseClass:
    # Propri√©t√©s et m√©thodes de la classe de base

class DerivedClass(BaseClass):
    # Propri√©t√©s et m√©thodes de la classe d√©riv√©e
```

Exemple d'h√©ritage simple:

```python
class FinantialAsset:
    EUR_TO_USD_RATE = 1.05
    def __init__(self, eur_value): # Constructeur, partag√© par tous les actifs financiers
        self.eur_value = eur_value
    def to_usd(self): # M√©thode partag√©e par tous les actifs financiers
        return self.eur_value * FinantialAsset.EUR_TO_USD_RATE

# Premier exemple sans __init__ dans la classe d√©riv√©e
class Stock(FinantialAsset):
    DIVIDEND_RATE = 0.03
    def compute_dividend(self): # M√©thode sp√©cifique aux actions
        return self.eur_value * Stock.DIVIDEND_RATE

class Bond(FinantialAsset):
    COUPON_RATE = 0.05
    def compute_coupon(self): # M√©thode sp√©cifique aux obligations
        return self.eur_value * Bond.COUPON_RATE

renault_stock = Stock(1000)
print(renault_stock.to_usd()) # 1050.0
print(renault_stock.compute_dividend()) # 30.0

french_bond = Bond(1000)
print(french_bond.to_usd()) # 1050.0
print(french_bond.compute_coupon()) # 50.0
```

## Surcharge de m√©thodes

Dans la plupart des cas, votre classe d√©riv√©e aura besoin de modifier ou d'√©tendre les m√©thodes de la classe de base. Pour cela, il suffit de red√©finir une m√©thode de la classe de base dans la classe d√©riv√©e. Cette technique est appel√©e **surcharge de m√©thode**.

```python
class BaseClass:
    def hello(self):
        print("Hello")
    def some_method(self):
        print("some_method from BaseClass")

class DerivedClass(BaseClass):
    def some_method(self):
        print("some_method from DerivedClass")

instance = DerivedClass()
instance.hello() # Hello
instance.some_method() # some_method from DerivedClass
```

## Extension des m√©thodes de la classe de base avec `super()`

Dans la plupart des cas nous allons vouloir modifier la m√©thode `__init__` de la classe de base dans la classe d√©riv√©e. Pour cela, nous pouvons utiliser la fonction `super()` qui permet d'appeler la m√©thode de la classe de base.

```python
class BaseClass:
    def __init__(self, value):
        self.value = value

class DerivedClass(BaseClass):
    def __init__(self, value, new_value):
        super().__init__(value) # Appel du constructeur de la classe de base
        self.new_value = new_value

instance = DerivedClass(10, 20)
print(instance.value) # 10
print(instance.new_value) # 20
```

Autre exemple :

```python
class FinancialAsset:
    EUR_TO_USD_RATE = 1.05
    def __init__(self, eur_value):
        self.eur_value = eur_value
    def to_usd(self):
        return self.eur_value * self.EUR_TO_USD_RATE

class Stock(FinancialAsset):
    def __init__(self, eur_value, dividend_rate):
        super().__init__(eur_value) # Appel du constructeur de la classe de base
        self.dividend_rate = dividend_rate # Ajout d'un nouvel attribut
    def compute_dividend(self):
        return self.eur_value * self.dividend_rate

renault_stock = Stock(1000, 0.03)
print(renault_stock.to_usd()) # 1050.0
print(renault_stock.compute_dividend()) # 30.0
```

> **Quand utiliser `super()`**
> - **Dans le constructeur de la classe d√©riv√©e**: pour appeler le constructeur de la classe de base
> - **Dans une m√©thode de la classe d√©riv√©e**: pour appeler une m√©thode de la classe de base

## V√©rifier l'appartenance d'une instance √† une classe avec `isinstance()`

La fonction `isinstance(some_instance, SomeClass)` permet de v√©rifier si une instance appartient √† une classe ou √† une classe d√©riv√©e.

```python
class BaseClass:
    pass

class DerivedClass(BaseClass):
    pass

class AnotherClass:
    pass

instance = DerivedClass()
print(isinstance(instance, DerivedClass)) # True
print(isinstance(instance, BaseClass)) # True
print(isinstance(instance, AnotherClass)) # False
```

> **üéä H√©ritage multiple** :
> Une classe d√©riv√©e peut h√©riter de plusieurs classes de base en utilisant la syntaxe suivante `class DerivedClass(BaseClass1, BaseClass2, ...):`.
> Il est r√©command√© d'√©viter l'h√©ritage multiple autant que possible, car cela peut rendre le code difficile √† comprendre et √† maintenir.


> **üéä Appeler un grand-parent sp√©cifique avec `super()`**: si vous avez plusieurs couches d'h√©ritage, vous pouvez appeler super √† partir d'un d'une classe sp√©cifique avec `super(ClasseFille, self)`. Ex :
> ```python
> class Parent:
>     def __init__(self, name):
>         self.name = name
> 
> class Child(Parent):
>     def __init__(self, name):
>         super().__init__(name+" Jr.")
> 
> class GrandChild(Child):
>     def __init__(self, name):
>         super(Child, self).__init__(name+" the second") # Appelle le constructeur `__init__` de la classe Parent
> 
> child = GrandChild('Donald')
> print(child.name) # Donald the second
> ```
